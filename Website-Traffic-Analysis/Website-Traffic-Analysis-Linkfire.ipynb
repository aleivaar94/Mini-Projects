# 4. [Pandas] What was the overall click rate (clicks/pageviews)?

```python
df.head()
```

```python
# filter only 'click' events and group them by the 'linkid' to get their count,
# i.e., number of clicks on each link
click_events = df[df['event'] == 'click'].groupby(['linkid', 'date'])['event'].size().reset_index()
click_events.rename(columns={'event': 'click_count'}, inplace=True)

page_events = df[df['event'] == 'pageview'].groupby(['linkid', 'date'])['event'].size().reset_index()
page_events.rename(columns={'event': 'page_count'}, inplace=True)
```

```python
click_rate = click_events.merge(right=page_events, on=['linkid', 'date'])
click_rate['click_rate'] = click_rate.click_count / click_rate.page_count
click_rate.sort_values('click_rate')
```

# 5. [Pandas] How does the clickrate distribute across different links?

```python
click_rate.describe()
```

Data is highly skewed due to the wide range of click_rate. Apply log transformation

```python
import numpy as np

click_rate['click_rate_log'] = np.log(click_rate.click_rate)
click_rate.sort_values('click_rate')
```

```python
import seaborn as sns

sns.boxplot(data=click_rate, x='click_rate_log')
```

```python
sns.displot(data=click_rate, x='click_rate_log');
```

# 6. [Pandas & SciPy] Is there any correlation between clicks and previews on a link? Is it significant? How large is the effect? Make sure to at least test for potential linear as well as categorical (think binary) relationships between both variables.

```python
# filter only 'click' events and group them by the 'linkid' to get their count,
# i.e., number of clicks on each link
link_clicks = df[df["event"] == "click"].groupby(by="linkid")["event"].count().reset_index()
link_clicks = link_clicks.rename(columns={"event": "clicks"})

# do the same for the 'preview' events
previews = df[df["event"] == "preview"].groupby(by="linkid")["event"].count().reset_index()
previews = previews.rename(columns={"event": "previews"})
```

```python
df_q6 = df[df["event"].isin(["click", "preview"])]
```

```python
# perform two database-style joins to append the number of link clicks and previews to each row
df_q6 = pd.merge(left=df_q6, right=link_clicks, how="left", on="linkid")
df_q6 = pd.merge(left=df_q6, right=previews, how="left", on="linkid")
```

```python
# drop null values
df_q6.dropna(subset=["clicks", "previews"], inplace=True)
df_q6.head()
```

Clicks and previews are positively correlated.

```python
df_q6.corr()
```

```python
from scipy import stats

stats.pearsonr(x=df_q6["clicks"], y=df_q6["previews"])
```

Using `pd.factorize()` to encode categorical variables into numeric representation.

There is a positive correlation between album, artist. (This is expected because there variables are naturally dependent, i.e., each track belongs to an album and is performed by an artist.)

```python
categorical_columns = ["country", "city", "artist", "album", "track"]

# [0] after pd.factorize returns the coded categorical variable
factorized_df = df_q6[categorical_columns].apply(lambda x: pd.factorize(x)[0])

# obtain correlation matrix
factorized_df.corr()
```

```python
# visualize correlation matrix
sns.heatmap(factorized_df, annot=True, cmap='Blues')
```
